```python
from tensorflow.keras.models import load_model
path='/content/drive/MyDrive/DNN'
model = load_model(f"{path}/model.h5")  # Adjust the filename if needed
```

```python

import cv2
import numpy as np
img = cv2.imread(f"{path}/download.jpg")  # Load image
img = cv2.resize(img, (48, 48))  # Resize to match model input
# Convert the image to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = img / 255.0  # Normalize if required
img = np.expand_dims(img, axis=0)  # Add batch dimension
img = np.expand_dims(img, axis=-1) # Add channel dimension
```

```python
# Define the class labels (adjust according to your dataset)
class_labels = {
    0: "Angry",
    1: "Disgust",
    2: "Fear",
    3: "Happy",
    4: "Neutral",
    5: "Sad",
    6: "Surprise"
}
predictions = model.predict(img)

# Get the predicted class index
expression_index = np.argmax(predictions)

# Get the corresponding class label
expression_label = class_labels.get(expression_index, "Unknown")

print(f"Predicted Expression: {expression_label}")
```
